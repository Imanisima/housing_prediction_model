```python
import pandas as pd
import numpy as np
import statistics
import matplotlib
%matplotlib inline
```

```python
# import housing dataset
data_path = "dataset/melb_data.csv"
housing_data = pd.read_csv(data_path)
```

```python
# let's get a glimpse at the data
housing_data.columns
```

### >> Observe <<
Notice how columns __BuildingArea__ and __CouncilArea__ have quite a few null values. Let's drop any missing (na) values from rows.

```python
housing_data.dropna(axis=0, inplace=True)
```

```python
housing_data.head()
```

```python
housing_data.tail()
```

### >> Choosing a Predition Target <<
Remember our goal is to predict the price of house prices. This is our target. We will call this (y).

```python
y = housing_data.Price
```

### >> Selecting Features <<
Next we want to choose features (or columns) needed to feed our model. We will call this list of features (x).

```python
# The most important features for determining price in our case will be Bathroom, Landsize, Location (Latitude, Longitude), Rooms.
melbourne_features = ["Rooms", "Bathroom", "Landsize", "Lattitude", "Longtitude"]
x = housing_data[melbourne_features]
```

```python
# let's get a summary of the feature data
x.describe()
```

```python
x.head()
```

## >> -- Build Model -- <<
Now we will create our model using scikit-learn library.

1. DEFINE the type of model
2. FIT patterns from the data
3. PREDICT
4. EVALUATE how accurate the model predictions are

```python
# necessary for creating models.
from sklearn.tree import DecisionTreeRegressor
```

```python
# define model
melbourne_model = DecisionTreeRegressor(random_state = 1)

# fit model using features, target variable
melbourne_model.fit(x, y)
```

```python
print("Predictions for the following 5 houses: ")
print(x.head())
```

```python
print("Price Predictions:")
print(melbourne_model.predict(x.head()))
```

```python
print("Actual price of above homes: ")
y.head()
```

## >> -- Model Validation -- <<

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

#### -> In-Sample <-

```python
insample_prediction = melbourne_model.predict(x)

mean_absolute_error(y, insample_prediction)
```

#### -> Out of Sample <-

```python
# randomly break up data into training and test data
# val_x = features
# val_y = target
train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 0)
```

```python
# define model
melbourne_model = DecisionTreeRegressor()

# fit model to train data
melbourne_model.fit(train_x, train_y)
```

```python
# get predicted prices on validation data
val_predictions = melbourne_model.predict(val_x)
print(mean_absolute_error(val_y, val_predictions))
```
