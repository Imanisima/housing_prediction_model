```python
import pandas as pd
import numpy as np
```

```python
iowa_data = pd.read_csv("dataset/iowa_data.csv", index_col='Id')
```

```python
# any null rows from with a missing target will be removed
iowa_data.dropna(axis = 0, subset=["SalePrice"], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# our target is the price of the house
y = iowa_data.SalePrice

# Separate the feature predictors from target
iowa_data.drop(["SalePrice"], axis = 1, inplace = True)
```

```python
iowa_data.info()
```

```python
# exclude categorial (object) variables
x = iowa_data.select_dtypes(exclude = ['object'])
```

```python
# separate data into test and train
x_train, x_val, y_train, y_val = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state = 0)
```

```python
# let's take a look at the train data set
x_train.head()
```

```python
# note that there are some missing values in the training set; denoted as 0.
x_train.shape
```

```python
# how many missing values in each column are there?
num_na_val_col = x_train.isnull().sum()
print(num_na_val_col[num_na_val_col > 0])
```
