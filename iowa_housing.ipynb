## >> -- Model Validation -- <<
### -> Out of Sample <-

```python
from sklearn.model_selection import train_test_split
```

```python
# split data into train-test
train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 1)
```

```python
# new model for sample verification
iowa_model = DecisionTreeRegressor(random_state = 1)

# Fit iowa_model with the training data.
iowa_model.fit(train_x, train_y)
```

```python
val_predictions = iowa_model.predict(val_x)
print(val_predictions)
```

```python
print(val_y.head())
```

```python
# calculate the mean absolute error in validation data
from sklearn.metrics import mean_absolute_error

val_mae = mean_absolute_error(val_predictions, val_y)
print("decision tree mae: ", val_mae)
```

Controlling overfitting/underfitting using MAE

```python
def calc_mae(max_leaf_nodes, train_x, val_x, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
    model.fit(train_x, train_y)
    
    predict_val = model.predict(val_x)
    mae = mean_absolute_error(predict_val, val_y)
    
    return mae
```

```python
# compare different tree sizes
possible_max_leaf_nodes = [5, 25, 50, 100, 250, 500]
for leaf in possible_max_leaf_nodes:
    mae = calc_mae(leaf, train_x, val_x, train_y, val_y)
    print("max leaf node: %d \t\t MAE: %d" %(leaf, mae))
```

The most optimal tree size: 100

### >> -- Final Model -- <<
Now that we have validated the data and found an adequate tree size for the model, let's use all of the data and keeping that tree size.

```python
max_tree_size = 100
final_iowa_model = DecisionTreeRegressor(max_leaf_nodes = max_tree_size, random_state = 1)
final_iowa_model.fit(x, y)
```

### >> -- Random Forest -- <<
To improve model performance, let's implement a random forest instead of a decision tree.

```python
from sklearn.ensemble import RandomForestRegressor

iowa_forest_model = RandomForestRegressor(random_state = 1)
iowa_forest_model.fit(train_x, train_y)

iowa_prediction = iowa_forest_model.predict(val_x)
print("random forest mae: ", mean_absolute_error(iowa_prediction, val_y))
```

Note: Compared to the decision tree's error of __29652__, the random forest's error is much better with a mae of __21857__!
